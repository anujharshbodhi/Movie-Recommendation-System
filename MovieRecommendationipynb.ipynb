```python
import pandas as pd
import numpy as np
```

```python
movie_df = pd.read_csv('moviedataset1.csv')
rating_df = pd.read_csv('ratings.csv')
```

```python
combine_movie_rating = pd.merge(rating_df, movie_df, on='movieId')
combine_movie_rating.head(10)
```

```python
columns = ['timestamp', 'genres']
combine_movie_rating = combine_movie_rating.drop(columns, axis=1)
combine_movie_rating.head(10)
```

```python
combine_movie_rating = combine_movie_rating.dropna(axis = 0, subset = ['title'])

movie_ratingCount = (combine_movie_rating.
     groupby(by = ['title'])['rating'].
     count().
     reset_index().
     rename(columns = {'rating': 'totalRatingCount'})
     [['title', 'totalRatingCount']]
    )
movie_ratingCount.head(10)
```

```python
rating_with_totalRatingCount = combine_movie_rating.merge(movie_ratingCount, left_on = 'title', right_on = 'title', how = 'left')
rating_with_totalRatingCount.head(10)
```

```python
user_rating = rating_with_totalRatingCount.drop_duplicates(['userId','title'])
user_rating.head(10)
```

```python
movie_user_rating_pivot = user_rating.pivot(index = 'userId', columns = 'title', values = 'rating').fillna(0)
movie_user_rating_pivot.head(10)
```

```python
X = movie_user_rating_pivot.values.T
X.shape
```

```python
import sklearn
from sklearn.decomposition import TruncatedSVD

SVD = TruncatedSVD(n_components=12, random_state=17)
matrix = SVD.fit_transform(X)
matrix.shape
```

```python
import warnings
warnings.filterwarnings("ignore",category =RuntimeWarning)
corr = np.corrcoef(matrix)
corr.shape
```

```python
# Assuming you already have your movie ratings dataset loaded and processed
# movie_user_rating_pivot should contain the user-item ratings matrix
# SVD should be your Singular Value Decomposition model

# Fit SVD to the user-item ratings matrix
X = movie_user_rating_pivot.values.T
SVD = TruncatedSVD(n_components=12, random_state=17)
matrix = SVD.fit_transform(X)
```

```python
# Compute cosine similarity
cosine_sim = cosine_similarity(matrix, matrix)
```

```python
# Define a function to get top recommendations for a given movie
def get_top_recommendations(movie_title, cosine_sim=cosine_sim, movie_user_rating_pivot=movie_user_rating_pivot):
    idx = movie_user_rating_pivot.columns.get_loc(movie_title)
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]  # Exclude the movie itself
    movie_indices = [i[0] for i in sim_scores]
    return movie_user_rating_pivot.columns[movie_indices]
```

```python
# Example: get top recommendations for a movie
movie_title = 'When Night Is Falling (1995)'
top_recommendations = get_top_recommendations(movie_title)
print("Top recommendations for", movie_title)
print(top_recommendations)
```
